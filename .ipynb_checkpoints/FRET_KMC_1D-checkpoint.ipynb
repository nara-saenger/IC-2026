```python
import random
import math
import matplotlib.pyplot as plt
import numpy as np
import json
from scipy.optimize import curve_fit
```

```python
plt.rcParams.update({
    'font.size': 14,
    'axes.titlesize': 17,
    'axes.labelsize': 17,
    'legend.fontsize': 12,
    'xtick.labelsize': 12,
    'ytick.labelsize': 12
})

plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'STIXGeneral'
```

```python
#Funções 

def calc_dt(k):
    return 1/sum(k)

def K_to_prob(k):
    K_T = sum(k)
    return [k[i]/K_T for i in range(len(k))]

def which_event(p):
    s = [sum(p[:i+1]) for i in range(len(p))]
    u = random.random()
    if u < s[0]:
        return 0
    if u < s[1]:
        return 1
    if u <= s[2]:
        return 2

def pulos(K,r):
    P = K_to_prob(K)
    dt = calc_dt(K)
    alive = True
    x = 0
    t = 0
    contagem_dir = 0
    contagem_esq = 0
    while(alive):
        event = which_event(P) 
        t += dt
        if event ==0:
            alive = False
        if event == 1:    #Pulo para a direita
            x += r 
            contagem_dir += 1
        if event ==2:     #Pulo para a esquerda 
            x -= r
            contagem_esq += 1
    return t, contagem_dir, contagem_esq    


```

```python
#Parâmetros 
R_F = 3
r = 1
Tau = 100
N =  10000     #Número de excitons

#Taxas
K_F = (R_F/r)**6/Tau      
K_Emi = 1/Tau

K = [K_Emi,K_F,K_F] 
```

```python
# Dados do tempo de emissão e contagem de pulos para N éxcitons

p = [pulos(K,r) for i in range(N)]
t_emi = [p[i][0] for i in range(N)]
pulos_dir = [p[i][1] for i in range(N)]
pulos_esq = [p[i][2] for i in range(N)]

dados = {"Tau_input": Tau,"t_emi": t_emi, "pulos_dir": pulos_dir, "pulos_esq": pulos_esq}
with open("dados_2.txt", "w") as arquivo:
    json.dump(dados, arquivo)
arquivo.close()

```

```python
# Histograma do tempo de emissão dos excitons

with open("dados_2.txt", "r") as arquivo:
    dados = json.load(arquivo)

t_emi = dados["t_emi"]
Tau_input = dados["Tau_input"]

Tau_calc = sum(t_emi)/len(t_emi)

#######################
import seaborn as sns

sns.histplot(t_emi, bins=30, kde=True)
plt.xlabel('Tempo de Emissão (s)')
plt.ylabel('Frequência')
plt.title('Histograma do Tempo de Emissão dos Excitons')
plt.show()
######################

def exp(t, A, Tau):
    return A * np.exp(-t/Tau)

y, bins = np.histogram(t_emi, bins=20, density=True)
bins = (bins[:-1] + bins[1:]) / 2

# remover zeros (importante pro log)
mask = y > 0
bins = bins[mask]
y = y[mask]

# log
log_y = np.log(y)

# ajuste linear
coef = np.polyfit(bins, log_y, 1)
a, b = coef  # a = inclinação, b = intercepto

Tau_linear = -1/a

# reta ajustada
y_linear_fit = a*bins + b

# plot
plt.figure()
plt.scatter(bins, log_y)
plt.plot(bins, y_linear_fit, color='red', label=f"τ = {Tau_linear:.2f}")
plt.xlabel("Emission time (a.u.)")
plt.ylabel("ln(Prob(t))")
plt.legend()

from matplotlib.ticker import MaxNLocator

plt.gca().xaxis.set_major_locator(MaxNLocator(5))  # no máx. 5 ticks no eixo x
plt.gca().yaxis.set_major_locator(MaxNLocator(5))
#plt.title("Linearização da distribuição exponencial")

plt.xlim(0,800)

#plt.title(f"Distruibuição do tempo de emissão")
plt.savefig(f"Conven_hist_lin.png")
plt.show()
plt.clf()

```

```python
# Arquivo para a animação 

def pulos_animacao(K,r):
    P = K_to_prob(K)
    dt = calc_dt(K)
    alive = True
    x = 0
    t = 0
    x_list = [x]
    t_list = [t] 
    while(alive):
        event = which_event(P) 
        t += dt
        if event ==0:
            alive = False
        if event == 1:    #Pulo para a direita
            x += r 
        if event ==2:     #Pulo para a esquerda 
            x -= r
        x_list.append(x)
        t_list.append(t)
    return x_list, t_list


X, T = pulos_animacao(K,r)
d = [(X[i],T[i]) for i in range(len(X))]

#print(d)

dados = {"lista animação": d}
with open("dados_animação_FRET1D.txt", "w") as arquivo:
    json.dump(dados, arquivo)
arquivo.close()
```
